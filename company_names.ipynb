```python
import yfinance
import pandas as pd
from urllib.error import HTTPError
```

```python
df = pd.read_csv('df_selected.csv')
```

```python
company_names = []
company_set = set()
for ticker in df["Ticker"]:
    ticker_obj = yfinance.Ticker(ticker)
    try: 
        long_name = ticker_obj.info.get('longName')
        print(ticker, long_name)
    except Exception as e:
        print(ticker, e)
        long_name = 'N/A'
    if long_name:
        if long_name not in company_set:
            company_names.append(long_name)
            company_set.add(long_name)
        else:
            company_names.append('N/A')
    else:
        company_names.append('N/A')
    # df.loc[df['Ticker'] == ticker, 'Company Name'] = long_name
```

```python
df["Company Name"] = company_names
```
